### Line Charts

- Line charts are traditionally used to visualize the development of a characteristic over time.
- We are going to test out a different type of line charts: **the development of the “prices” column across the days of the week**
    - To achieve this, we’d plot the “orders_day_of_week” column on the x-axis and the “prices” column on the y-axis.
    - caveat: this line chart will require a great deal of computing power due to the number of unique values in the “prices” column.
    - To avoid this, we can use the subsetting method - sampling data.

#### sampling data
- create a subset that actually represents the entirety of our data.  
- more importantly, those rows would need to be randomized.
- Steps:
    1. Set the seed for the random number generator.
    2. Create 32,404,859 random numbers (the length of your `ords_prods_merge` dataframe) ranging between 0 and 1.
    3. Code a logical test comparing each of those numbers to a value of 0.7 (either less than or greater than 0.7).
    4. Store the results of the test in a new list, `dev`, in the form of boolean values (`True` and `False`).
    5. split your dataframe into two samples: big, small
    6. Check if the number of rows adds up correctly

```python
# Creating a sample from the dataframe
# value is 0.7, since we want a 70/30 split. 
# create a new list called `dev`: it will be what stores our boolean values - we’ll be getting an output of `True` or `False`

np.random.seed(4)
dev = np.random.rand(len(df_exercise)) <= 0.7 
```

```python
dev
```

```python
# Create a list holding True/False values to the test np.random.rant() <= 0.7

dev = np.random.rand(len(df_exercise)) <= 0.7
```

```python
# Store 70% of the sample in the dataframe "big"

big = df_exercise[dev]
```

```python
# Store 30% of the sample in the dataframe "small"
# The tilde negates the argument, it stores everything that’s not dev, or False.

small = df_exercise[~dev] 
```

```python
# Check if the number of rows adds up correctly

len(df_exercise)
```

```python
len(big) + len(small)
```

- Since the numbers match, we can assume that our subsetting procedure was successful.
- One last step before creating that chart:
    - to speed up the processing time even further, we'll reduce our samples to *only* those columns necessary for the chart:
        - “prices” and “orders_day_of_week.

```python
# by creating a new dataframe, df_2, that contains only those columns necessary for the chart: “prices” and “orders_day_of_week” columns from our small subset.

df_2 = small[['orders_day_of_week', 'prices']]
```

Now we are ready to create the line chart:

```python
# Create the line chart using sns.lineplot() function (from the seaborn library) and save it to a new object: line. 

line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

```python
# Export the line chart using the "figure.savefig()" function

line.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_prices_orders_day_of_week.png'))
```

4. Store the results of the test in a new list, in the form of boolean values (True and False).
